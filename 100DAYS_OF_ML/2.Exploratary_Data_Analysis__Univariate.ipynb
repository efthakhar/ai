# Categorical Data Analysis: Univariate

```python
import seaborn as sns
import pandas as pd
```

### Count and Plot the frequency of Each Category of a Categorical Column

```python
titanic_df = pd.read_csv('data/titanic.csv')
#sns.countplot(titanic_df['Sex'])
sns.countplot(titanic_df['Embarked'])
```

```python
titanic_df['Sex'].value_counts().plot(kind='bar')
```

```python
titanic_df['Survived'].value_counts().plot(kind='bar')
```

```python
titanic_df['Survived'].value_counts().plot(kind='pie', autopct='%.3f')
```

# Numerical Data Analysis: Univariate

## 1. Histogram: Graphical representation of the distribution of a dataset

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(titanic_df['Age'], bins=10)
```

```python
sns.displot(titanic_df['Age'], kind='hist', kde=True)
```

## 2. Boxplot

```python
sns.boxplot(titanic_df['Age'])
```
